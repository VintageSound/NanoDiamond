```python
import os

os.chdir(r"D:\NanoDiamond\PulseSequencer")

from Interfaces.dataSaver import dataSaver
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from IPython.display import display
```

```python
dataSaver = dataSaver() 
```

```python
def plotLoadedData(data: pd.DataFrame):
    x_label = data.columns[0]
    y_label = data.columns[1]
    
    plt.plot(data[x_label], data[y_label])
    plt.xlabel = x_label
    plt.ylabel = y_label

    plt.show()
```

# Load ODMR Files

```python
directory_path_ODMR = r"D:\Experiments\2023-12-18\ODMR"
```

```python
files = os.listdir(directory_path_ODMR)

for file in files:
    print(file)
    
    path = os.path.join(directory_path_ODMR, file)
    pdMetadata, data = dataSaver.loadPickle(path)
    
    display(pdMetadata)
    display(data)
    plotLoadedData(data)
```
